```python
dict_df = df.to_dict('records')
dict_df
total_list = []
for items in dict_df:
    temp_list = list(items.values())
    total_list.append(temp_list)
#     print(new_list)
# print(total_list)
# d = {"name":"python", "version":3.9}
# new_list = list(dict_df.values())
# print(new_list)

from nltk import flatten
import math

flattened_list = flatten(total_list)
# print(flattened_list)

c_list = [x for x in flattened_list if pd.isnull(x) == False]
# print(c_list)

counts = {item:c_list.count(item) for item in c_list}
print(counts)


import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize =(10, 50), dpi = 80)

keys = counts.keys()
values = counts.values()
x_pos = [i for i, _ in enumerate(keys)]

plt.style.use('ggplot')

plt.barh(x_pos, values, color='green')
plt.yticks(x_pos, keys)

plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.3', color='red')

plt.show()
```
